# Display processed sensor data

This notebook displays the processed sensor data and comments on some of the observed patterns.

General info on the tracer experiment is in the [`readme.md`](../readme.md) file in the home folder. Details about the sensors and data pre-processing are in the [`README_processed_sensor_data.txt`](README_processed_sensor_data.txt) file.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read raw data
v = '2023-12-05'
T = pd.read_csv(f'processed_logger_data_{v}.csv')
input_time = pd.to_datetime('12-Sep-2023 16:40') #time when first tracer application started

# Convert date column to datetime
T['date'] = pd.to_datetime(T['date'])
T['day'] = (T['date'].apply(pd.to_datetime,dayfirst=True) - input_time) / np.timedelta64(1, 'D') #days after the tracer application

# Report useful dates
in_dates = pd.to_datetime(['12-Sep-2023 16:43', '13-Sep-2023 17:30'])  # Tracer input dates UTC+01
in_days = (in_dates - input_time) / np.timedelta64(1, 'D') #days after the tracer application
download_dates = pd.to_datetime(['13-Sep-2023 08:28', '14-Sep-2023 07:58', '23-Sep-2023 14:12'])  # Logger download dates UTC+01
download_days = (download_dates - input_time) / np.timedelta64(1, 'D') #days after the tracer application
```

```python
# VWC
rows, cols = 4, 1
fig, axs = plt.subplots(rows, cols, figsize=(12, 12))
colors = ['tab:blue','tab:orange','tab:green','tab:red']
ymin,ymax = [0, 0.4]

# depth 10
axs[0].set_ylim([ymin, ymax])
axs[0].set_ylabel('VWC [-]')
axs[0].set_xlabel('days after tracer application [d]')
axs[0].set_title('Depth 10 cm')

axs[0].axvline(x=in_days[0], linestyle='-', color='.6', label='tracer application 1')
axs[0].axvline(x=in_days[1], linestyle='--', color='.6', label='tracer application 2')

axs[0].plot(T['day'], T['VWC_L_10'], label='VWC_L_10',color=colors[0])
axs[0].plot(T['day'], T['VWC_M_10'], label='VWC_M_10',color=colors[1])
axs[0].plot(T['day'], T['VWC_R_10'], label='VWC_R_10',color=colors[2])

axs[0].legend()

# depth 20
axs[1].set_ylim([ymin, ymax])
axs[1].set_ylabel('VWC [-]')
axs[1].set_xlabel('days after tracer application [d]')
axs[1].set_title('Depth 20 cm')

axs[1].axvline(x=in_days[0], linestyle='-', color='.6', label='_tracer application 1')
axs[1].axvline(x=in_days[1], linestyle='--', color='.6', label='_tracer application 2')

axs[1].plot(T['day'], T['VWC_L_20'], label='VWC_L_20',color=colors[0])
axs[1].plot(T['day'], T['VWC_M_20'], label='VWC_M_20',color=colors[1])
axs[1].plot(T['day'], T['VWC_R_20'], label='VWC_R_20',color=colors[2])

axs[1].legend()

# depth 30
axs[2].set_ylim([ymin, ymax])
axs[2].set_ylabel('VWC [-]')
axs[2].set_xlabel('days after tracer application [d]')
axs[2].set_title('Depth 30 cm')

axs[2].axvline(x=in_days[0], linestyle='-', color='.6', label='_tracer application 1')
axs[2].axvline(x=in_days[1], linestyle='--', color='.6', label='_tracer application 2')

axs[2].plot(T['day'], T['VWC_L_30'], label='VWC_L_30',color=colors[0])
axs[2].plot(T['day'], T['VWC_M_30'], label='VWC_M_30',color=colors[1])

axs[2].legend()

# depth 40
axs[3].set_ylim([ymin, ymax])
axs[3].set_ylabel('VWC [-]')
axs[3].set_xlabel('days after tracer application [d]')
axs[3].set_title('Depth 40 cm')

axs[3].axvline(x=in_days[0], linestyle='-', color='.6', label='_tracer application 1')
axs[3].axvline(x=in_days[1], linestyle='--', color='.6', label='_tracer application 2')

axs[3].plot(T['day'], T['VWC_M_40'], label='VWC_M_40',color=colors[1])

axs[3].legend()

plt.tight_layout()
plt.show()


# EC
rows, cols = 4, 1
fig, axs = plt.subplots(rows, cols, figsize=(14, 12))
colors = ['tab:blue','tab:orange','tab:green','tab:red']
ymin,ymax = [0, 3000]

# depth 10
axs[0].set_ylim([ymin, ymax])
axs[0].set_ylabel('EC [uS/cm]')
axs[0].set_xlabel('days after tracer application [d]')
axs[0].set_title('Depth 10 cm')

axs[0].axvline(x=in_days[0], linestyle='-', color='.6', label='tracer application 1')
axs[0].axvline(x=in_days[1], linestyle='--', color='.6', label='tracer application 2')

axs[0].plot(T['day'], T['EC_L_10'], label='EC_L_10',color=colors[0])
axs[0].plot(T['day'], T['EC_M_10'], label='EC_M_10',color=colors[1])
axs[0].plot(T['day'], T['EC_R_10'], label='EC_R_10',color=colors[2])

axs[0].legend()

# depth 20
axs[1].set_ylim([ymin, ymax])
axs[1].set_ylabel('EC [uS/cm]')
axs[1].set_xlabel('days after tracer application [d]')
axs[1].set_title('Depth 20 cm')

axs[1].axvline(x=in_days[0], linestyle='-', color='.6', label='_tracer application 1')
axs[1].axvline(x=in_days[1], linestyle='--', color='.6', label='_tracer application 2')

axs[1].plot(T['day'], T['EC_L_20'], label='EC_L_20',color=colors[0])
axs[1].plot(T['day'], T['EC_M_20'], label='EC_M_20',color=colors[1])
axs[1].plot(T['day'], T['EC_R_20'], label='EC_R_20',color=colors[2])

axs[1].legend()

# depth 30
axs[2].set_ylim([ymin, ymax])
axs[2].set_ylabel('EC [uS/cm]')
axs[2].set_xlabel('days after tracer application [d]')
axs[2].set_title('Depth 30 cm')

axs[2].axvline(x=in_days[0], linestyle='-', color='.6', label='_tracer application 1')
axs[2].axvline(x=in_days[1], linestyle='--', color='.6', label='_tracer application 2')

axs[2].plot(T['day'], T['EC_L_30'], label='EC_L_30',color=colors[0])
axs[2].plot(T['day'], T['EC_M_30'], label='EC_M_30',color=colors[1])

axs[2].legend()

# depth 40
axs[3].set_ylim([ymin, ymax])
axs[3].set_ylabel('EC [uS/cm]')
axs[3].set_xlabel('days after tracer application [d]')
axs[3].set_title('Depth 40 cm')

axs[3].axvline(x=in_days[0], linestyle='-', color='.6', label='_tracer application 1')
axs[3].axvline(x=in_days[1], linestyle='--', color='.6', label='_tracer application 2')

axs[3].plot(T['day'], T['EC_M_40'], label='EC_M_40',color=colors[1])

axs[3].legend()

plt.tight_layout()
plt.show()
```

### What we can observe

The breakthrough following the first tracer application is almost invisible because the soil was extremely dry and the sensor was likely not in the ideal conditions to record reliable data. Moreover, measurements were heavily disturbed by animals excavating around the sensors area.

The breakthrough following the second tracer application was well recorded in the VWC and EC data. While the response is very heterogeneous, we can clearly see that sensors at 10 cm depth responded immediately, sensors at 20 cm depth responded more slowly and at 30 and 40 cm depth there was no VWC nor tracer response
